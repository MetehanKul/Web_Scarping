```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import time

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36 OPR/113.0.0.0"
}

a = 1
liste = []

while a <= 50:
    try:
        r = requests.get(f"https://www.hepsiburada.com/ara?q=gamer&sayfa="+str(a)+"", headers=headers)
        soup = BeautifulSoup(r.content, "lxml")

        st1 = soup.find("div", attrs={"class": "productListContent-tEA_8hfkPU5pDSjuFdKG"})
        st2 = st1.find("ul", attrs={"class": "productListContent-frGrtf5XrVXRwJ05HUfU productListContent-rEYj2_8SETJUeqNhyzSm"})
        st3 = st2.find_all("li", attrs={"class": "productListContent-zAP0Y5msy8OHn5z7T_K_"})

        for i in st3:
            link_end = i.a.get("href")
            link_start = "https://www.hepsiburada.com/"
            link = link_start + link_end
            print(link)

            r1 = requests.get(link, headers=headers)
            soup1 = BeautifulSoup(r1.content, "lxml")
            time.sleep(1)

            try:
                Title = soup1.find("h1", attrs={"class": "xeL9CQ3JILmYoQPCgDcl"}).text.strip()
                Salesman = soup1.find("span", attrs={"class": "rzVCX6O5Vz9bkKB61N2W"}).text.strip()
                Evaluation = soup1.find("a", attrs={"class": "yPPu6UogPlaotjhx1Qki"}).text.strip()
                Rating = soup1.find("span", attrs={"class": "JYHIcZ8Z_Gz7VXzxFB96"}).text.strip()
                Price = soup1.find("div", attrs={"data-test-id": "price-current-price"}).text.strip()
            except:
                Title = np.nan
                Salesman = np.nan
                Rating = np.nan
                Evaluation = np.nan
                Price = np.nan

            liste.append([Title, Salesman, Rating, Evaluation, Price])
        
        a += 1
        time.sleep(2)
    except Exception as e:
        print(f"Error on page {a}: {e}")
        a += 1
        time.sleep(2)

df = pd.DataFrame(liste, columns=["Title", "Salesman", "Rating", "Evaluation", "Price"])
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.to_excel("hepsiburada_gamer.xlsx", index=False)
```
